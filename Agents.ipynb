```python
from langchain_classic.agents import initialize_agent, load_tools, AgentType
```

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model

# Load the environment variables from the .env file
load_dotenv()

# Debug: confirm the key was loaded
api_key = os.getenv("GROQ_API_KEY")
print("Key found:", api_key is not None)
print("Key prefix:", api_key[:7] if api_key else "None")
```

```python
# Delete this cell or replace with:
from langchain_groq import ChatGroq
llm = ChatGroq(model="llama-3.3-70b-versatile", temperature=0.9)
```

```python
tools = load_tools(["llm-math", "wikipedia"], llm=llm)
```

```python
agent = initialize_agent(tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True)
```

```python
agent.run("What is the 25% of 200?")
```
